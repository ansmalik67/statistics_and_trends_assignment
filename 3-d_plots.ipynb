
### **3D ploting**

3D plotting is a graphical technique that uses X, Y, and Z axes to visualize three-dimensional data points in a coordinate system

```python
def plot_3d_scatter(data, x, y, z, hue=None, title=None):
    """
    Create a 3D scatter plot.

    Parameters:
    data (DataFrame): The DataFrame containing the data.
    x (str): The name of the column for the x-axis.
    y (str): The name of the column for the y-axis.
    z (str): The name of the column for the z-axis.
    hue (str): The name of the column to use for color encoding (categorical).
    title (str): The title of the plot.
    """
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    if hue:
        scatter = ax.scatter(data[x], data[y], data[z],
                             c=data[hue].astype('category').cat.codes, cmap='viridis', marker='o')
        legend1 = ax.legend(*scatter.legend_elements(), title=hue)
        ax.add_artist(legend1)
    else:
        ax.scatter(data[x], data[y], data[z], color='b', marker='o')

    ax.set_xlabel(x)
    ax.set_ylabel(y)
    ax.set_zlabel(z)
    ax.set_title(title if title else f'3D Scatter Plot of {x}, {y}, {z}',  pad=30, color="Green", fontsize=20)
    ax.view_init(elev=30, azim=120)

    plt.show()

# 3d scatter plot with salaries, training levels and sales
plot_3d_scatter(data=data, x='salary', y='training level', z='sales', hue='level of education')
```

### **Comparison Plots**

Comparison plotting is a data visualization method that allows users to compare statistics across populations and data sets

```python
# joint plot graph
sns.jointplot(x='work experience',y='salary',data=data)
```

### **PairPlot**

A pair plot, also known as a scatterplot matrix, is a matrix of graphs that enables the visualization of the relationship between each pair of variables in a dataset.

```python
# pairplot data graph
sns.pairplot(data)
```

### **catplot()**

A function in the Seaborn library that creates categorical plots on a FacetGrid
# Kind Options are: “point”, “bar”, “strip”, “swarm”, “box”, “violin”, or “boxen”

```python
# catplot graph with whole data
sns.catplot(x='work experience',y='salary',data=data,kind='swarm',col='level of education',row='division')
plt.xticks(rotation=45);
```
